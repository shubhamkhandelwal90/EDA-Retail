## 2.3 Data Cleaning

### 2.3.1 Renaming Data Columns

```python
# Checking Columns

terror_data.columns.values
```

```python
# Renaming Columns

terror_data.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','gname':'Group','country_txt':'Country',
                            'region_txt':'Region','provstate':'State','city':'City',
                            'latitude':'Latitude', 'longitude':'Longitude','attacktype1_txt':'AttackType',
                            'targtype1_txt':'TargetType','weaptype1_txt':'Weapon','nkill':'Kill',
                             'nwound':'Wound','target1':'Target1'},inplace=True)
```

```python
terror_data = terror_data[['Year','Month','Day','Country','State','Region','City','Latitude','Longitude',"AttackType",'Kill',
               'Wound','Target1','Group','TargetType','Weapon']]
```

```python
# Checking shape of new data

terror_data.shape
```

```python
terror_data.info()
```

### 2.3.2 Null Value Treatment

```python
# Checking Null Values

terror_data.isnull().sum()
```

```python
# Fixing Null Values
terror_data['State']=terror_data['State'].fillna('Others')
terror_data['City']=terror_data['City'].fillna('Others')
terror_data['Target1']=terror_data['Target1'].fillna('Others')
terror_data['TargetType']=terror_data['TargetType'].fillna('Others')
terror_data['Wound'] = terror_data['Wound'].fillna(0)
terror_data['Kill'] = terror_data['Kill'].fillna(0)
terror_data['Latitude'].fillna(int(terror_data['Latitude'].mean()), inplace=True)
terror_data['Longitude'].fillna(int(terror_data['Longitude'].mean()), inplace=True)
```

```python
terror_data.isnull().sum()
```

```python
# Creating new column as casaualties

terror_data['Casualties'] = terror_data['Wound'] + terror_data['Kill']
```

```python
terror_data.Weapon.unique()
```

```python
terror_data.Month.unique()
```

```python
# Formating the Column Value
terror_data.replace( {"Vehicle (not to include vehicle-borne explosives, i.e., car or truck bombs)": "Vehicle"},inplace = True)
terror_data.replace( {0: 6},inplace = True)
```

```python
terror_data.Weapon.unique()
```

```python
terror_data.Month.unique()
```
