# 3. Visualizing the Data

## 3.1 Correlation in the Data

```python
cor = store_data.corr()
plt.figure(figsize = (8,5))
plt.title('Correaltion among other variables',fontdict={'fontsize':20,'fontweight':5,'color':'green'})
sns.heatmap(cor, annot = True)
plt.show()
```

```python
store_data1 = store_data.corr().abs().unstack().sort_values(ascending=False).drop_duplicates()
store_data1
```

## 3.2 Column Wise Data Analysis

```python
plt.figure(figsize=(8,5))
sns.set_style('whitegrid') 
sns.countplot(store_data['Ship Mode'],order = store_data['Ship Mode'].value_counts().index,palette='Set2')
plt.title('Top Shipping Mode',fontdict={'fontsize':20,'fontweight':5,'color':'green'})
plt.ylabel("Count", fontdict={'fontsize':20,'fontweight':15,'color':'green'})
plt.xlabel("Shipping Mode", fontdict={'fontsize':20,'fontweight':15,'color':'green'})
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.set_style('whitegrid') 
sns.countplot(x = store_data['Segment'], order = store_data['Segment'].value_counts().index,palette=['#432371',"#FAAE7B"])
plt.title('Top Segment',fontdict={'fontsize':20,'fontweight':5,'color':'green'})
plt.ylabel("Count", fontdict={'fontsize':20,'fontweight':15,'color':'green'})
plt.xlabel("Segment", fontdict={'fontsize':20,'fontweight':15,'color':'green'})
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.set_style('whitegrid') 
sns.countplot(x = store_data['Category'], order = store_data['Category'].value_counts().index, palette='Set1')
plt.title('Top Categories',fontdict={'fontsize':20,'fontweight':5,'color':'green'})
plt.ylabel("Count", fontdict={'fontsize':20,'fontweight':15,'color':'green'})
plt.xlabel("Category", fontdict={'fontsize':20,'fontweight':15,'color':'green'})
plt.show()
```

```python
plt.figure(figsize=(20,9))
plt.xticks(rotation=90)
sns.set_style('whitegrid') 
sns.countplot(x = store_data['Sub-Category'], order = store_data['Sub-Category'].value_counts().index,palette='spring')
plt.title('Top Sub-Categories',fontdict={'fontsize':20,'fontweight':5,'color':'green'})
plt.ylabel("Count", fontdict={'fontsize':20,'fontweight':15,'color':'green'})
plt.xlabel("Sub-Category", fontdict={'fontsize':20,'fontweight':15,'color':'green'})
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.set_style('whitegrid') 
sns.countplot(store_data['Region'],order = store_data['Region'].value_counts().index,palette='Set3')
plt.title('Top Region',fontdict={'fontsize':20,'fontweight':5,'color':'green'})
plt.ylabel("Count", fontdict={'fontsize':20,'fontweight':15,'color':'green'})
plt.xlabel("Region", fontdict={'fontsize':20,'fontweight':15,'color':'green'})
plt.show()
```

```python
plt.figure(figsize=(20,9))
plt.xticks(rotation=90)
sns.set_style('whitegrid') 
sns.countplot(store_data['State'],order = store_data['State'].value_counts().index,palette=["#fc9272","#fee0d2"])
plt.title('Top States',fontdict={'fontsize':20,'fontweight':5,'color':'green'})
plt.ylabel("Count", fontdict={'fontsize':20,'fontweight':15,'color':'green'})
plt.xlabel("State", fontdict={'fontsize':20,'fontweight':15,'color':'green'})
plt.show()
```

```python
plt.figure(figsize = (20,9))
sns.countplot((store_data['City']), order=store_data['City'].value_counts().iloc[:31].index,palette=["#fee0d2","#7fcdbb","#edf8b1"])
plt.xticks(rotation = 90)
plt.title('Top Cities',fontdict={'fontsize':20,'fontweight':5,'color':'green'})
plt.ylabel("Count", fontdict={'fontsize':20,'fontweight':15,'color':'green'})
plt.xlabel("City", fontdict={'fontsize':20,'fontweight':15,'color':'green'})
plt.show()
```

## 3.3 Sales,Profit,Discount Analysis of Data

### 3.3.1 Ship Mode Wise

```python
store_data.groupby('Ship Mode')['Sales','Discount','Profit'].mean().plot(kind='pie', autopct='%1.1f%%',figsize=(20,10),
                       subplots=True,textprops={'fontsize': 14},legend=True,
                                                                       colors = ['purple','gold','pink','lightgreen'])

plt.title('Ship Mode Wise Sales-Discount-Profit', fontdict={'fontsize':20,'fontweight':5,'color':'green'})
plt.show()
```

### 3.3.2 Segment Wise

```python
store_data.groupby('Segment')['Sales','Discount','Profit'].mean().plot(kind='pie', autopct='%1.1f%%',figsize=(20,10),
                            startangle=6,subplots=True,textprops={'fontsize': 14})

plt.title('Segment Wise Sales-Discount-Profit', fontdict={'fontsize':20,'fontweight':5,'color':'green'})
plt.show()
```

### 3.3.3 Category Wise

```python
store_data.groupby('Category')['Sales','Discount','Profit'].mean().plot(kind='pie', autopct='%1.1f%%',figsize=(20,10),
                           subplots=True,textprops={'fontsize': 14},
                                                                        colors=['pink','skyblue','brown'])

plt.title('Category Wise Sales-Discount-Profit', fontdict={'fontsize':20,'fontweight':5,'color':'green'},loc='center')
plt.show()
```

### 3.3.4 Sub-Category Wise Analysis

#### A. Sales Wise

```python
plt.title('Sub-Category Wise Sales', fontdict={'fontsize':20,'fontweight':5,'color':'green'})
store_data.groupby(['Sub-Category'])['Sales'].mean().plot(kind='pie', autopct='%1.1f%%',
                                    figsize=(20,60))
plt.show()
```

#### B. Discount

```python
plt.title('Sub-Category Wise Discount', fontdict={'fontsize':20,'fontweight':5,'color':'green'})
store_data.groupby(['Sub-Category'])['Discount'].mean().plot(kind='pie', autopct='%0.2f%%',figsize=(20,60))
plt.show()
```

#### C. Profit

```python
store_data.groupby(['Sub-Category'])['Profit'].mean().plot(kind='bar',figsize=(15,4))
plt.title('Sub-Category Wise Profit', fontdict={'fontsize':20,'fontweight':5,'color':'green'})
plt.show()
```

### 3.3.5  Region Wise Sales

```python
store_data.groupby('Region')['Sales','Discount','Profit'].mean().plot(kind='pie', autopct='%0.2f%%',figsize=(18,20),
                            startangle=60,subplots=True,legend=True,textprops={'fontsize': 14},
                                                                       colors=['aqua','red','yellow','blue'])

plt.title('Region Wise Sales-Discount-Profit', fontdict={'fontsize':20,'fontweight':5,'color':'green'})
plt.show()
```

### 3.3.5 State Wise

#### A. Sales

```python
plt.title('State Wise Sales', fontdict={'fontsize':20,'fontweight':5,'color':'green'})
store_data.groupby(['State'])['Sales'].mean().plot(kind='pie', autopct='%1.1f%%',textprops={'fontsize': 12},
                                startangle=90,figsize=(20,70))
plt.show()
```

#### B. Discount

```python
plt.title('State Wise Discount', fontdict={'fontsize':20,'fontweight':5,'color':'green'})
store_data.groupby('State')['Discount'].mean().plot(kind='bar',figsize=(15,4),color='red',legend=True)
plt.ylabel("Discount", fontdict={'fontsize':20,'fontweight':15,'color':'green'})
plt.xlabel("State", fontdict={'fontsize':20,'fontweight':15,'color':'green'})
plt.show()
```

#### C. Profit

```python
store_data.groupby('State')['Profit'].mean().plot(kind='bar',figsize=(15,4),legend=True,color='skyblue')
plt.title('State Wise Profit', fontdict={'fontsize':20,'fontweight':5,'color':'green'})
plt.ylabel("Profit", fontdict={'fontsize':20,'fontweight':15,'color':'green'})
plt.xlabel("State", fontdict={'fontsize':20,'fontweight':15,'color':'green'})
plt.show()
```

### 3.3.6 City Wise Analysis

#### A. Positive Sales & Positive Profit

```python
store_data.groupby('City')['Sales','Profit'].mean().head(30)
```

#### B. Positive Sales & Negative Profit

```python
store_data.groupby('City')['Sales','Profit'].mean().sort_values(by='Profit',ascending = False).tail(30)
```

```python
store_data.groupby('City')['Sales','Profit'].mean().sort_values(by='Profit',ascending = False).tail(30).plot(kind='bar',figsize=(15,4),color=['yellow','purple'])
plt.title('City With Positive Sales & Negative Profit', fontdict={'fontsize':20,'fontweight':5,'color':'green'})
plt.ylabel("Sales", fontdict={'fontsize':20,'fontweight':15,'color':'green'})
plt.xlabel("City", fontdict={'fontsize':20,'fontweight':15,'color':'green'})
plt.show()
```

#### C. Discount

```python
plt.title('City Wise Discount', fontdict={'fontsize':20,'fontweight':5,'color':'green'})
store_data.groupby('City')['Discount'].mean().sort_values(ascending = False).head(30).plot(kind='bar',figsize=(15,4),color='brown')
plt.ylabel("Discount", fontdict={'fontsize':20,'fontweight':15,'color':'green'})
plt.xlabel("City", fontdict={'fontsize':20,'fontweight':15,'color':'green'})
plt.show()
```
